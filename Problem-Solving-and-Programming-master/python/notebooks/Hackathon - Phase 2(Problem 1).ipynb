### Problem Statement :
#### Using the Matplotlib library, generate a Pie Chart to categorise student marks data in a CSV file into Distinction, First Class, Second Class, Third Class and Failed Categories(based on percentages). Explode the Distinction category.
 
#### (a)Generate a CSV containing marks of 2000 students

```python
import random,pandas
def addMarks(marksfile,no_ofstudents):
    df=pandas.read_csv(marksfile)
    for i in range(2210410001,no_ofstudents+2210410001):
        df.loc[len(df)]=[i,random.randrange(0,101)]
    df.to_csv(marksfile,index=False)
    return df

addMarks('DataFiles/Marks.csv',2000)
```

#### (b)From the CSV, generate the Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

def distinction(n):
    return n>=80
def first(n):
    return 60<=n<80
def second(n):
    return 50<=n<60

df=pd.read_csv('DataFiles/Marks.csv',usecols=['Marks'])
marks=list(map(int,df.values))
dis=(sum(map(distinction,marks)))/2000
fi=(sum(map(first,marks)))/2000
s=(sum(map(second,marks)))/2000
t=(sum(1 for i in marks if i>=40 and i<50))/2000
fail=(sum(1 for i in marks if i<40))/2000
labels='Distinction','First class','Second class','Third class','Fail'
sizes=[dis*100,fi*100,s*100,t*100,fail*100]
colors=['green','blue','orange','red','yellow']
explode=(0.1,0,0,0,0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.2f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```
