### Problem 2: 
#### Problem Statement :
Plotting a sine graph 

```python
import numpy as np

t=np.arange(0.0,2.0,0.01)
v=1+np.sin(2*np.pi*t)
plt.xlabel('time')
plt.ylabel('voltage')
plt.title('sine wave')
plt.plot(t,v)
plt.grid(True)
plt.savefig('Figures/time-voltage.png')
plt.show()
```

### Problem 3: 
#### Problem Statement :
Plot a line graph on **income vs year** for the state of **california**

```python
import pandas as pd

df=pd.read_csv('DataFiles/Income.csv')
y=df.values[4]
x=df.columns
years=[]
for i in range(2,len(x)):
    years.append(int(x[i]))
income=[]
for i in range(2,len(y)):
    income.append(y[i])
plt.xlabel('year')
plt.ylabel('income')
plt.title('California State')
plt.grid()
plt.plot(years,income)
plt.savefig('Figures/line-graph-income.png')
```

```python

width=1/2
plt.xlabel('year')
plt.ylabel('income')
plt.title('California State')
plt.bar(years,income,width,color="Green")
plt.savefig('Figures/bar-graph-income.png')
plt.show()
```

### Problem 4: 
#### Problem Statement :
Generate a scatter plot for 50 random numbers

```python
n=50
x=np.random.rand(n)
y=np.random.rand(n)
colorlist=np.random.rand(n)
arealist=np.pi*(15*np.random.rand(50))**2
plt.scatter(x,y,s=arealist,c=colorlist,alpha=1)
plt.savefig('Figures/scatterplot-50randomnumbers.png')
```
